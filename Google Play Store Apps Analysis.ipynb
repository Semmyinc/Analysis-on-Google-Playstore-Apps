```python
# total number of app titles that contain 'astrology'
filt_astrology = df['App'].str.contains('astrology', case=False)
len(df[filt_astrology])
```

```python
df.isnull().sum()
percentage_missing_entries = df.isnull().sum()*100/len(df)
percentage_missing_entries
```

```python
# find average app rating
df['Rating'].mean()
```

```python
# find total number of unique category
df['Category'].nunique()
```

```python
# which category is getting the highest average rating
df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```

```python
# find total number of apps having 5 star rating
filt_app_rating = df['Rating'] == 5
len(df[filt_app_rating]['App'])
```

```python
# find average value of reviews
# df['Reviews'] = pd.to_numeric(df['Reviews'])
# filt = df['Reviews'] == '3.0M'
# df[filt]
df['Reviews'] = df['Reviews'].replace('3.0M', 3.0)
df['Reviews']
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'])
# df['Reviews'].dtypes
df['Reviews'].mean()
```

```python
# total number of free and paid apps
df.Type.value_counts()
```

```python
# which app has max reviews?
# mtd 1
df.groupby('App')['Reviews'].max().sort_values(ascending=False)
```

```python
# mtd 2
filt_max_reviews = df['Reviews'] == df['Reviews'].max() 
df[filt_max_reviews]['App']
```

```python
# Display top 5 apps having highest reviews
df.groupby('App')['Reviews'].max().sort_values(ascending=False).head()
```

```python
# find average rating of free and paid apps
df.groupby('Type')['Rating'].mean()
```
